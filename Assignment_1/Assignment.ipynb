```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
df_google = pd.read_csv("GOOG.csv")
closing_prices = df_google["close"].values

window_size = 100
rolling_window = np.ones(window_size) / window_size
smoothed_prices = np.convolve(closing_prices, rolling_window, mode="same")

plt.figure(figsize=(16, 8))
plt.plot(df_google["date"], smoothed_prices, label="Smoothed Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Google Stock Price Analysis")
plt.legend()
plt.show()
```
