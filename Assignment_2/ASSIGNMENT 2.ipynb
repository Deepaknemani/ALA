```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
patient1 = [76,126,38.0]
patient2 =[74,120,38.0]
patient3 = [72,118,37.5]
patient4 = [78,136,37.0]
```

```python
patients = [patient1, patient2, patient3, patient4]
data = np.array(patients)
print(patients)
```

```python
import numpy as np
distances = np.linalg.norm(data[:, np.newaxis] - data, axis=-1)
farthest_patient_idx = np.unravel_index(np.argmax(distances), distances.shape)
nearest_patient_idx = np.unravel_index(np.argmin(distances), distances.shape)
print("The patient with index", farthest_patient_idx[0], "is farthest from patient with index", farthest_patient_idx[1], "with a distance of", distances[farthest_patient_idx])
```

```python
import numpy as np
distances = np.linalg.norm(data[:, np.newaxis] - data, axis=-1)
np.fill_diagonal(distances, np.inf)
nearest_patient_indices = np.unravel_index(np.argmin(distances), distances.shape)
patient1_index, patient2_index = nearest_patient_indices
min_distance = distances[patient1_index, patient2_index]
print("The two nearest patients are Patient", patient1_index + 1, "and Patient", patient2_index + 1)
print("Distance:", min_distance)
```

```python
import numpy as np
distances = np.linalg.norm(data[:, np.newaxis] - data, axis=-1)
new_dummy_patient = np.array([75, 130, 37.8])
new_patient_distances = np.linalg.norm(data - new_dummy_patient, axis=1)
closest_patient_index = np.argmin(new_patient_distances)
print("The closest patient to the new dummy patient is Patient", closest_patient_index + 1)
print("Distance:", new_patient_distances[closest_patient_index])
```
